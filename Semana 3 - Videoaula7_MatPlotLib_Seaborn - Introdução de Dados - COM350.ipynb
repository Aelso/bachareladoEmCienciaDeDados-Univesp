#Gerar um novo Gráfico (para valor de Y sendo o quadrado do valor de X)

A função **arange** criar um sequência de numeros indicada nesse caso de começando de **-100** até **99**. Indo de **1 em 1**.

```python
x1=np.arange(-100,100,1)
```

**Mostrando** os **valores** de **x1**.

```python
x1
```

Como **não** temos os valores de **y** aqui. Os valores de **y** será **(X^2)**. Temos então um **gráfico da função de 2º grau** é **representado** pela **parábola**.




```python
plt.plot(x1,x1**2)
plt.show()
```

#Experimentando ver cortar o eixo X (vamos diminuir um pouco o valor de Y)

Os valores de y será **x quadrado** subtraído **2000**.
##Exemplo: **(x,y)**.
##(2, (2^2-2000)) = **(2,-1996)** 

```python
plt.plot(x1,(x1**2)-2000)
plt.show()
```

#Mais uma experiência com dados fabricados

A função **arange** simulando um mês com 30 dias.Começando pelo dia **primeiro**  até dia **30**. 

```python
dias=np.arange(1,31)
dias
```

Explicando a variável **vacinados**: Gerar valores de **0** até **1000** esses valores será gerado 30 vezes, ou seja, simula o números de vacinado durante **um mês**. A variável **contagios** terá mesmo comportamento variável **vacinados** porém o valor máximo será de **700**.

```python
vacinados=np.random.randint(0, 1000,30)
contagios=np.random.randint(0,700,30)
```

```python
vacinados
```

```python
contagios
```

#Plotar Gráfico

Abaixo temos algumas possibilidade de plotar o gráfico.Usamos o comando **plt.style.use** e indicamos qual tipo de gráfico queremos.
##link de referência dos gráficos disponível na biblioteca matplotlib.
 https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

```python
#plt.style.use('classic')
#plt.style.use('dark_background')
plt.style.use('default')

plt.figure(figsize=(10,5))
plt.bar(dias,vacinados)
plt.plot(dias,contagios,'r')
plt.ylabel('Vacinados por Dia')
plt.show()
```

#Criar Dataframe com os dados fabricados

Podemos criar um **dataframe** com dados.

```python
dados=pd.DataFrame(dias, columns=['Dias'])
```

```python
dados['Contagios']=contagios
dados['Vacinados']=vacinados
```

```python
dados
```

#Plotar gráfico a partir de um DF

Gerar um gráfico usando dados do dataframe com recursos do panda.
**plotando** um gráfico com **plot** do tipo **barra(bar)**, onde **x** será **dias** e **y** vacinados. **Observer** que usa a biblioteca matplotlib para gerar o gráfico.
###**matplotlib.axes._subplots.AxesSubplot at 0x7fd6593d5d90>**


```python
dados.plot(kind='bar', x='Dias', y='Vacinados')
```

#Experimentar a Biblioteca Seaborn

Gerar um gráfico usando dados do dataframe com recursos do **seaborn**.
**plotando** um gráfico do tipo barra **barplot**, onde **x** será **Dias** e **y** **Contagios**. 
Esse receberá um barra em **vermelho (color='r')** no eixo **y** com informações dos **Vacinados**.
###**sns.barplot(data=dados, x='Dias', y='Vacinados', color='r')**
 
##**Observer** que usa a biblioteca matplotlib para gerar o gráfico.



```python
sns.barplot(data=dados, x='Dias', y='Contagios')
sns.barplot(data=dados, x='Dias', y='Vacinados', color='r')
```

**Outra forma** de **visualizar** o mesmo gráfico. Plotando os mesmos dados só que dessa vez um **linha vermelho** (color='r') no eixo y com informações dos **Vacinados**.

```python
sns.barplot(data=dados, x='Dias', y='Contagios')
sns.lineplot(data=dados, x='Dias', y='Vacinados', color='r')
```
